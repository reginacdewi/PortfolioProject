```python
#libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #to adjust plot chart size
```

```python
#read the data

df = pd.read_csv('/Users/regina.dewi/Downloads/movies.csv')
```

```python
#check the data content
df.head()
```

## Clean and setup the data

```python
#check for missing data

## not using formats
# for col in df.columns:
#     pct_missing = np.mean(df[col].isnull())
#     print(col, pct_missing) 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    


```

```python
#check amount of missing values 
missing_values = df.isnull().sum()
print(missing_values)

print(df.shape)
```

```python
#drop missing values, use thresh = 2 with threshold at least 2 missing values only, but in here I will delete all rows with missing value
#axis = 'index' means that I only want to delete rows, not the column
#inplace = true means that I want to modify the original df, if inplace=false (the default) then it'll not be returned to the modified vers

df.dropna(axis='index', inplace = True)

#check the data after removing null values
print(df.shape)
```

```python
#check data types of column
df.dtypes
```

```python
#change data type of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df.head()
```

```python
#sort the data by gross value, and save it
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Drop duplicates
#if I want to drop all duplicates should use df.drop_duplicates()
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#display max rows (all rows)

pd.set_option('display.max_rows', None)
```

## Finding The Correlations
Hypothesis: budget and company will have high correlation to gross income?


```python
#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Budget')
plt.ylabel('Gross Earning')
plt.show()
```

```python
#plot it using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#check correlation data

df.corr(method='pearson') #only works for numerical data, pearson correlation method is the default
#there are pearson, kendall, and spearman correlation

```

```python
#create correlation heatmap
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numerical Variables')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#check for non-numerical variables

#first, convert the object variable to categorical

df_num = df


# Create a new DataFrame to store the mapping
mapping_df = pd.DataFrame(columns=['column', 'code'])

for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object'):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes
    
        
df_num.head()
```

```python
#create correlation heatmap
correlation_matrix = df_num.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numerical Variables')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_num.corr()
corr_pairs =  correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending=False)
sorted_pairs
```
