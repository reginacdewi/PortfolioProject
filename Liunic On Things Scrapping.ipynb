## Append the data

```python
#append all dataframe
liunic_tokopedia = pd.concat([df_bag, df_iphone, df_apparel,df_stationary], axis=0)

#reset the index
liunic_tokopedia = liunic_tokopedia.reset_index(drop=True)

#change data to integer
liunic_tokopedia['sold'] = liunic_tokopedia['sold'].str.replace('+', '').astype(int)
liunic_tokopedia['product_price'] = liunic_tokopedia['product_price'].str.replace('.', '').astype(int)

liunic_tokopedia['est_revenue'] = liunic_tokopedia['product_price']*liunic_tokopedia['sold']

liunic_tokopedia.sort_values(by=['est_revenue'], inplace=False, ascending=False)
```
